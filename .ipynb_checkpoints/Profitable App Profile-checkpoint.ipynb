Above we see that Instagram appears 4 times in the app. I wonder how many duplicate apps we have in the store. 

```python
duplicate_apps = []
unique_apps = []
for _, app in android_ds.iterrows():
    name = app['App']
    if name in unique_apps:
        duplicate_apps.append(name)
    else:
        unique_apps.append(name)
        
print('Number of duplicate apps:', len(duplicate_apps), '\n')
print('Examples of duplicate apps:', duplicate_apps[:10])
```

Let's make sure that there really are duplicats of some of the sample apps above. We'll use 'Google My Business', 'ZOOM Cloud Meetings', and 'Slack

```python
gmb = 'Google My Business'
zoom = 'ZOOM Cloud Meetings'
slack = 'Slack'
dupes = [gmb, zoom, slack]

for _, row in android_ds.iterrows():
    if row['App'] in dupes:
        print(row, '\n')
        
```

Instead of simply removing duplicates at random, we'll use the number of reviews to decide which row for duplicate apps to keep in our data set. This makes intuitive sense since users tend to be drawn toward the apps with the highest number of reviews.

```python
# create a dict to hold our duplicate apps and their number of reviews
reviews_max = {}
for idx, row in android_ds.iterrows():
    name = row['App']
    n_reviews = float(row['Reviews']) 
    
    if reviews_max.has_key(name) and review_max[name] < n_reviews:
        review_max[name] = n_reviews
    
    if name not in reviews_max:
        reviews_max[name] = n_reviews
```
