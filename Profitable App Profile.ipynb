Instead of simply removing duplicates at random, we'll use the number of reviews to decide which row for duplicate apps to keep in our data set. This makes intuitive sense since users tend to be drawn toward the apps with the highest number of reviews.

````python
# create a dict to hold our duplicate apps and their number of reviews
reviews_max = {}
for idx, row in android_ds.iterrows():
    name = row['App']
    n_reviews = float(row['Reviews']) 
    
    if name in reviews_max and reviews_max[name] < n_reviews:
        reviews_max[name] = n_reviews
    
    if name not in reviews_max:
        reviews_max[name] = n_reviews
````

Next, let's define remove any apps not intended for an English-speaking audience. We can use Python's ```ord()``` function to tranpose a character into a given integer, and we know that english-characters are in the range of 0-127. So we'll write a function to determine whether a given app name is english-oriented or not. We'll return ```True``` or ```False``` accordingly and use that to filter out rowns from our dataframe in order to retain an English-oriented dataset.

Additionally, we'll make sure that we're only filtering out app names in our function if there are more than 4 non-english characters. This is because the 0-127 range above only captures the most standard of english characters, including lowercase and uppercase letters, but not emjois or symbols like "™". While we could create a more elaborate function to deal with these use cases, we'll keep it simple for now. The end result should still be fairly effective without requiring a much more involved function.

````python
def is_english_word(app_name):
    non_english_count = 0
    for char in app_name:
        char_int = ord(char)
        if char_int > 127:
            non_english_count += 1
            if non_english_count > 4:
                return False
    return True
````

Next, we'll use the 'Genres' and 'Category' columns in the android dataset and the 'prime_genre' category in the ios dataset to determine which types of apps tend to be the most popular. We'll do this by creating a frequency table for each of those categories.

We'll define a function to do this for each of the genres we're interested in. The function will take in a dataset, and the column name we're interested in. It'll create a dictionary with the the discrete column values as the key and the count of apps under that column. Since dictionaries aren't all that useful for sorting, we'll create a list of tuples (each tuple being a key-value pair from that dictionary) and sort the list. This sorted tuple list is the frequency table that our function will return.

````python
def return_freq_table(dataset, col_name):
    categories = {}
    for _, row in dataset.iterrows():
        category = row[col_name]
        if category in categories:
            categories[category] += 1
        else:
            categories[category] = 1
    
    freq_table = []
    for k in categories:
        freq_tuple = (categories[k], k)
        freq_table.append(freq_tuple)
    
    freq_table = sorted(freq_table)
    return freq_table
````

````python
def print_frequency_table(list_of_tuples):
    for tup in list_of_tuples:
        print(tup[1], ':', tup[0])
````
